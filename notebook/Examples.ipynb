```python
from Bio import SeqIO
from seqann import BioSeqAnn
from BioSQL import BioSeqDatabase
```

```python
# Initialize BioSQL Database
server = BioSeqDatabase.open_database(driver="pymysql", user="root",
                                      passwd="my-secret-pw", host="localhost",
                                      db="bioseqdb", port=3307)
```

```python
seqann = BioSeqAnn(server=server, verbose=True, verbosity=5, pid="example-1/2")
input_file = 'exact_seqs.fasta'
for seq in SeqIO.parse(input_file, "fasta"):
    ann = seqann.annotate(seq)
    for feat in ann.annotation:
        print(feat, ann.annotation[feat], sep="\t")
```
